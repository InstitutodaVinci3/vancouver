```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('https://raw.githubusercontent.com/InstitutodaVinci3/Vancouver/master/Experimento/TEMPERATURA_VAGINAL.csv', sep=';', encoding='latin1', engine='python')
df.head()
```

```
df.describe()
```

```
df['TEMPERATURA_ANIMAL'] = df['TEMPERATURA_ANIMAL'].astype(str) 
df['TEMPERATURA_ANIMAL'] = df['TEMPERATURA_ANIMAL'].apply(lambda x:x.replace(",",".")) #replace , por .
df['TEMPERATURA_ANIMAL'] = df['TEMPERATURA_ANIMAL'].astype(float) #converte toda a coluna para float
```

```
df.dtypes
```

```
 df['TEMPERATURA_ANIMAL'].hist(bins=80)
```

```
plt.figure(figsize=(90,8))
sns.countplot(data=df,x='TEMPERATURA_ANIMAL')
```
