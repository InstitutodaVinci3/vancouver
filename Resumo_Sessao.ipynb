```
path =r'/content/drive/MyDrive/Colab Notebooks/MedMep DataSet/Resumo_Sessao CSV'
filenames = glob.glob(path + "/*.csv")
```

```
Resumo_df = pd.DataFrame({'Id_Animal':[],'Grupo':[], 'Dias_Lac':[], 'Num_Lac':[],  'Data':[], 'Nome_Sala':[], 'Num_Sala':[],
                          'Sessao_#':[], 'Lote_#':[], 'MPC':[], 'Hora_ID':[], 'Hora_I':[], 'Duracao':[], 'Hora_F':[], 'Producao':[], 
                          'Prod_Relativa':[], '2_min':[], 'LEC':[], 'Baixa_Prod':[], '2a_Volta':[]})  
for filename in filenames:
  a = pd.read_csv(filename, skiprows=1, skipfooter=1, sep=',', encoding='latin1', engine='python',    
                  names=['Id_Animal','Grupo', 'Dias_Lac', 'Num_Lac',  'Data', 'Nome_Sala', 'Num_Sala', 'Sessao_#', 
                         'Lote_#', 'MPC', 'Hora_ID', 'Hora_I', 'Duracao', 'Hora_F', 'Producao', 
                         'Prod_Relativa', '2_min', 'LEC', 'Baixa_Prod', '2a_Volta'])
  Resumo_df = pd.concat([Resumo_df, a])
  Resumo_df['Data_Arq'] = filename  # Ainda não está OK - está ficando o nome do último arquivo somente
```
